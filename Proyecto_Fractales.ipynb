Primer fractal

```python
from pylab import *
import math
#Método que recibe dos puntos y dibuja una linea úniendolos
def drawline(pi,pf): #pi, pf son son puntos de forma p=[x,y]
    px=[pi[0],pf[0]]
    py=[pi[1],pf[1]]
    plot(px,py, color='purple')
    xlim(0,10)
    ylim(0,7)
    
def met(pcen,iter,tam): #(punto del centro, iteraciones, tamaño de línea)
    if iter>=1:
        p1=(pcen[0],pcen[1]-(tam*math.cos(3*math.pi)))
        p2=(pcen[0]-(tam*math.cos(.06981)),pcen[1]-(tam*math.sin(.6981)))
        p3=(pcen[0]+(tam*math.cos(.06981)),pcen[1]-(tam*math.sin(.6981)))
        drawline(pcen,p1)
        drawline(pcen,p2)
        drawline(pcen,p3)
        if iter>1:
            met(p1,iter-1,tam*.5)
            met(p2,iter-1,tam*.5)
            met(p3,iter-1,tam*.5)
    
a=int(input("Iteraciones:"))
met([5,3],a,2) #(punto central, iteraciones, tamaño de los brazos) 
```

Segundo Fractal

```python
def drawline(pi,pf): #pi, pf son son puntos de forma p=[x,y]
    px=[pi[0],pf[0]]
    py=[pi[1],pf[1]]
    plot(px,py, color='limegreen')
    xlim(0,7)
    ylim(0,5)
def drawrect(px,py,a,b): #(esquina inicial, ancho, largo)
    p1=[px,py]
    p2=[px+a,py]
    p3=[px+a,py+b]
    p4=[px,py+b]
    drawline(p1,p2)
    drawline(p2,p3)
    drawline(p3,p4)
    drawline(p4,p1)
def fracRect(nivel, size,x,y):#nivel, tamaño, coordenada en x, coordenada en y
    if nivel==1:
        drawrect(2,1.5,2,2)
    else:
        drawrect(x + ((size*.25) + (size*.5)), y - (size*.25), size*.5, size*.5)
        drawrect(x - (size*.25), y - (size*.25), size*.5, size*.5)
        drawrect(x- (size*.25), y + ((size*.25) + size*.5), size*.5, size*.5)
        drawrect(x + ((size*.25) + size*.5), y + ((size*.25) + size*.5), size*.5, size*.5)
        
        fracRect(nivel - 1, size*.5, x + ((size*.25) + size*.5), y + ((size*.25) + size*.5))
        fracRect(nivel - 1 , size*.5 , x + (size*.25) + (size*.5), y - (size*.25))
        fracRect(nivel -1, size*.5, x - (size*.25), y - (size*.25))
        fracRect(nivel -1, size*.5, x - (size*.25), y + ((size*.25) + size*.5))
a=int(input("Iteraciones:"))
fracRect(a,2,2,1.5)#nivel, tamaño, coordenada en x, coordenada en y
                 
```

Tercer Fractal. Triángulo de Sierpinski

```python
def drawline(pi,pf): #pi, pf son son puntos de forma p=[x,y]
    px=[pi[0],pf[0]]
    py=[pi[1],pf[1]]
    plot(px,py, color='limegreen')
    xlim(0,7)
    ylim(0,5)

def fracSir(nivel, size,x,y):#nivel, tamaño, coordenada en x, coordenada en y
    if nivel==1:
        drawrect(2,1.5,2,2)
    else:
        drawrect(x + ((size*.25) + (size*.5)), y - (size*.25), size*.5, size*.5)
        drawrect(x - (size*.25), y - (size*.25), size*.5, size*.5)
        drawrect(x- (size*.25), y + ((size*.25) + size*.5), size*.5, size*.5)
        drawrect(x + ((size*.25) + size*.5), y + ((size*.25) + size*.5), size*.5, size*.5)
        
        fracRect(nivel - 1, size*.5, x + ((size*.25) + size*.5), y + ((size*.25) + size*.5))
        fracRect(nivel - 1 , size*.5 , x + (size*.25) + (size*.5), y - (size*.25))
        fracRect(nivel -1, size*.5, x - (size*.25), y - (size*.25))
        fracRect(nivel -1, size*.5, x - (size*.25), y + ((size*.25) + size*.5))
a=int(input("Iteraciones:"))
fracRect(a,2,2,1.5)#nivel, tamaño, coordenada en x, coordenada en y
```

```python
def drawline(pi,pf): #pi, pf son son puntos de forma p=[x,y]
    px=[pi[0],pf[0]]
    py=[pi[1],pf[1]]
    plot(px,py, color='Royalblue')
    xlim(0,7)
    ylim(0,5)
def drawtriangle(pA,pD,pI): #Punto de arriba, derecha e izquierda
    drawline(pA,pD)
    drawline(pD,pI)
    drawline(pI,pA)
def midPoint(p1,p2): #Punto de arriba, derecha e izquierda
    P=[(p1[0]+p2[0])/2,(p1[1]+p2[1])/2]
    return P
def fracSir(pA,pI,pD, iter):#Punto de arriba, izquierda y derecha, iteraciones
    if iter==1:
        drawtriangle(pA,pD,pI)
    else:
        pMidFondo=midPoint(pD,pI)
        pMidDerecha=midPoint(pA,pD)
        pMidIzquierda=midPoint(pA,pI)
        
        fracSir(pA,pMidIzquierda,pMidDerecha,iter-1)
        fracSir(pMidIzquierda,pI,pMidFondo,iter-1)
        fracSir(pMidDerecha,pMidFondo,pD,iter-1)
a=int(input("Iteraciones:"))
fracSir([1,1],[5,1],[3,4.6],a)
```

Cuarto Fractal. Curva de Koch

```python
def drawline(pi,pf): #pi, pf son son puntos de forma p=[x,y]
    px=[pi[0],pf[0]]
    py=[pi[1],pf[1]]
    plot(px,py, color='Crimson')
    xlim(0,7)
    ylim(0,5)
def trPoint(p1,p2,t): #dos puntos, una razon 
    #encuentra a dos puntos entre dos dados
    P=[(p1[0]*(1-t))+(p2[0]*t),(p1[1]*(1-t))+(p2[1]*t)]
    return P
def rotar(p,cen,angle):#rotar un punto
    pa=[p[0]-cen[0],p[1]-cen[1]]#mover al origen
    pb=[pa[0]*math.cos(angle)-pa[1]*math.sin(angle),pa[0]*math.sin(angle)+pa[1]*math.cos(angle)]#rotar
    pr=[pb[0]+cen[0],pb[1]+cen[1]]#regresar el punto
    return pr
def koch(n,p0,p4,angle):#iteraciones, punto inicial, ángulo en radianes, longitud
    p=[p0,p0,p0,p0,p4]
    p[1]=trPoint(p[0],p[4],1/3)
    p[3]=trPoint(p[0],p[4],2/3)
    p[2]=rotar(p[3],p[1],angle)
    i=0
    while i<4:
        if n<=1:
            drawline(p[i],p[i+1])
        else:1
            koch(n-1,p[i],p[i+1],angle)
        i+=1
a=int(input("Iteraciones:"))
koch(a,[.5,1],[6.5,1],math.pi/3)
```

Conjunto de Mandelbrot (ilustrativo)

```python
#este método es meramente ilustrativo, no fue realizado por el equipo
import matplotlib.pyplot as plt
import numpy as np
from numba.roc.decorators import autojit

def mandelbrot(Re, Im, max_iter):
    c=complex(Re, Im)
    z=0.0j
    
    for i in range(max_iter):
        z=z*z+c
        if(z.real*z.real+z.imag*z.imag) >= 4:
            return i
    
    return max_iter

columns=2000
rows=2000

result= np.zeros([rows, columns])
for row_index, Re in enumerate(np.linspace(-2, 1, num=rows)):
    for column_index, Im in enumerate(np.linspace(-1, 1, num=columns)):
        result[row_index, column_index]=mandelbrot(Re, Im, 100)
        
plt.figure(dpi=100)
plt.imshow(result.T, cmap='hot', interpolation='bilinear', extent=[-2,1,-1,1])
plt.xlabel('Re')
plt.ylabel('Im')
plt.show()
```
